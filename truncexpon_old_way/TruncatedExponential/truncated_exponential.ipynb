# Truncated Exponential

The truncated exponential is defined as:
\begin{equation}
f(x,b) = \frac{e^{-x}}{1-e^{-b}}
\end{equation}

for $x \in [0,b]$.

## Scipy implementation

```python
from scipy.stats import truncexpon
import matplotlib.pyplot as plt
import numpy as np
```

## Plotting by picking random values from the distribution

```python
# truncexp parameters 
b = 2
loc = 0
scale = 1.75

# Set the bins
bins = np.linspace(loc,loc+b,100)

# Generate pdf and array of random variates
x = np.linspace(loc-0.3,loc+b+0.3,1000)
pdf = truncexpon.pdf(x,b=b/scale,loc=loc,scale=scale)
pdf_anti = truncexpon.pdf(x,b=b/scale,loc=loc,scale=scale)[::-1]
r_worm = truncexpon.rvs(b=b/scale,loc=loc,scale=scale,size=int(5E+05))
r_anti = -truncexpon.rvs(b=b/scale,loc=loc,scale=scale,size=int(5E+05))+b

# Plot histogram and compare with pdf
fig, ax = plt.subplots(1, 1)
ax.plot(x,pdf,label=r"$\Delta\tau_{worm}$",color='blue')
ax.plot(x,pdf_anti,label=r"$\Delta\tau_{anti}$",color='orange')
ax.hist(r_worm,bins=bins,density=True,histtype='stepfilled',alpha=0.45,color='blue') ;
ax.hist(r_anti,bins=bins,density=True,histtype='stepfilled',alpha=0.45,color='orange') ;
plt.legend(loc='best');
ax.set_ylabel(r'$counts$')
ax.set_xlabel(r'$\Delta\tau$')
plt.savefig('flipped_truncated_exponential.pdf')
```
