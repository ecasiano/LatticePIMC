```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data
file_name = "nloc_enloc.dat"
data = np.loadtxt(file_name)
```

```python
# Plot nloc vs mu/U
plt.plot(data[:,0],data[:,1],label=r'$n_{loc}$',ls='--')
plt.plot(data[:,0],data[:,2],label=r'$\epsilon_{n_{loc}}$',ls='-')
plt.xlabel(r'$\mu/U$')
plt.legend()
plt.savefig("nloc_enloc.pdf")
```
