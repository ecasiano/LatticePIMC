```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data
file_name = "nloc_enloc_U_1000000.dat"
data = np.loadtxt(file_name)
```

```python
# Calculate exact \eps_{loc}
mu_list = np.linspace(-1,1.75,1000)
eps_list = np.zeros_like(mu_list)
for i,mu in enumerate(mu_list):
    if mu < 0:
        n_i = 0
    else:
        n_i = int(mu+1)
    eps_list[i] = (1/2)*n_i*(n_i-1)-mu*n_i
```

```python
# Plot nloc vs mu/U
plt.plot(data[:,0],data[:,1],label=r'$n_{loc}$',ls='--')
plt.plot(data[:,0],data[:,2],label=r'$\epsilon_{n_{loc}}/U$',ls=':',lw=4,zorder=2)
plt.plot(mu_list,eps_list,label=r'$\frac{1}{2}n_{\rm{loc}}(n_{\rm{loc}}-1))-\frac{\mu}{U} n_{\rm{loc}}$',ls='-',zorder=1)

plt.xlabel(r'$\mu/U$')
plt.legend()
plt.savefig("nloc_enloc_U_1000000.pdf")
```
